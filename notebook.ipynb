### Cleaning Data

Instant dan windspeed tidak dibutuhkan pada data. maka kita bersihkan terlebih dahulu

```python
df.drop(columns=['instant','windspeed'],inplace=True)
df.head()
```

Ubah kolom menjadi kategorikal

Mengganti nama kolom

```python
df.rename(columns={
    'dteday':'datetime',
    'yr':'year',
    'mnth':'month',
    'hr':'hour',
    'weathersit':'weather_situation',
    'cnt':'count'
},inplace=True)
df.head()
```

Mengubah tipe data kolom datetime ke tipe data yang sesuai

```python
df['datetime'] = pd.to_datetime(df['datetime'])
df.info()
```

Mengubah kolom menjadi kategorikal

```python
df['season'] = df['season'].map({
    1: 'Winter', 2: 'Spring', 3: 'Summer', 4: 'Fall'
})
df.head()
```

```python
df['year'] = df['datetime'].dt.year
df.head()
```

```python
df['month'] = df['datetime'].dt.month_name()
df.head()
```

```python
df['weekday'] = df['datetime'].dt.day_name()
df.head()
```

```python
df['weather_situation'] = df['weather_situation'].map({
    1: 'Clear',
    2: 'Cloudy',
    3: 'Light Snow/Rain',
    4: 'Heavy Snow/Rain'
})
df.head()
```

```python
df.info()
```

pada info diatas banyak data yang tidak sesuai.

Merubah data-data dibawah menjadi tipedata kategorikal

```python
df['season'] = df['season'].astype('category')
df['year'] = df['year'].astype('category')
df['month'] = df['month'].astype('category')
df['weekday'] = df['weekday'].astype('category')
df['hour'] = df['hour'].astype('category')
df['holiday'] = df['holiday'].astype('category')
df['workingday'] = df['workingday'].astype('category')
df['weather_situation'] = df['weather_situation'].astype('category')
df.info()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Pada pukul berapa masyarakat merental sepeda?



```python
hour_df = df.groupby(by='hour').agg({
    'count':'mean'
}).reset_index()
hour_df.head()
```

```python
plt.figure(figsize=(10,5))
# sns.boxplot(data=hour_df,x='hour',y='count')
plt.plot(hour_df['hour'],hour_df['count'],marker='o', color='red')
plt.title("Penyewaan sepeda berdasarkan jam")
plt.xticks(hour_df['hour'])
plt.show()
```

### Pertanyaan 2: Bagaimana kondisi cuaca berdampak pada tingkat penyewaan sepeda?

```python
weathersit_df = df.groupby(by='weather_situation').agg({
    'count':'sum'
}).reset_index()
weathersit_df
```

```python
plt.figure(figsize=(10,5))
plt.bar(weathersit_df['weather_situation'], weathersit_df['count'], color='orange')
plt.title('Weather Situation Counts')
plt.show()
```

```python
temp_df = df.groupby(by='weather_situation').agg({
    'temp':'mean',
    'atemp':'mean',
    'hum':'mean',
}).reset_index()
temp_df
```
